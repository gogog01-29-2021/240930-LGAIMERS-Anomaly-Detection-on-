```python
# 'OK' 값을 NaN으로 변환하고 결측치 처리할 열 리스트
columns_to_nan = [
    'HEAD NORMAL COORDINATE X AXIS(Stage1) Collect Result_Dam',
    'HEAD NORMAL COORDINATE X AXIS(Stage1) Collect Result_Fill1',
    'HEAD NORMAL COORDINATE X AXIS(Stage1) Collect Result_Fill2'
]
```

```python
#최빈값으로 설정
def preprocess_coordinates(df, columns_to_replace):
    for column in columns_to_replace:
        df[column].replace('OK', np.nan, inplace=True)
    
    # 숫자로 변환할 수 없는 데이터를 NaN으로 변환
    for column in columns_to_replace:
        df[column] = pd.to_numeric(df[column], errors='raise')
    
    for column in columns_to_replace:
        mode_value = df[column].mode()[0]
        df[column].fillna(mode_value, inplace=True)
    
    return df
```

```python
#범주형과 수치형분리
categorical_cols = x.select_dtypes(include=['object', 'category']).columns
numerical_cols = x.select_dtypes(include=['number']).columns

# 범주형 변수의 쌍을 찾는 함수
def find_categorical_relationships(df):
    relationships = {}
    for col in categorical_cols:
        unique_values = df[col].unique()  #칼럼중에 고유값찾고
        for value in unique_values:
            subset = df[df[col] == value]
            if len(subset) > 1:
                for other_col in categorical_cols:
                    if other_col != col:
                        unique_other_values = subset[other_col].unique() #고유값이 다 똑같다면 relationship변수
                        if len(unique_other_values) == 1:
                            relationships[col] = other_col
    return relationships

# 수치형 변수의 쌍을 찾는 함수
def find_numerical_relationships(df):
    relationships = {}
    for col in numerical_cols:
        grouped = df.groupby(col).size()
        if grouped.size < len(df):
            for other_col in numerical_cols:
                if other_col != col:
                    if df.groupby(col)[other_col].nunique().max() == 1:
                        relationships[col] = other_col
    return relationships

# 관계 찾기
cat_relationships = find_categorical_relationships(x)
num_relationships = find_numerical_relationships(x)
```

```python

# 대표 변수 선택
def select_representative_vars(relationships):
    selected_vars = set()
    all_vars = set()  # 전체 선택된 변수들을 기록하기 위한 집합

    for key, value in relationships.items():
        if key not in all_vars and value not in all_vars:
            selected_vars.add(key)
            selected_vars.add(value)
            all_vars.add(key)
            all_vars.add(value)
    
    # 중복되는 변수 제거
    final_selected_vars = set()
    for var in selected_vars:
        if not any(var in rel for rel in relationships.values()):
            final_selected_vars.add(var)
    
    return list(final_selected_vars)

selected_categorical_cols = select_representative_vars(cat_relationships)
selected_numerical_cols = select_representative_vars(num_relationships)
```

```python
import os
# StratifiedKFold 설정
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)

os.makedirs('renew_smote_tomek', exist_ok=True)

# 각 폴드에 대한 데이터를 저장할 리스트
resampled_X_train_list = []
resampled_y_train_list = []
X_test_list = []
y_test_list = []

# K-fold 교차 검증 수행
for fold, (train_index, test_index) in enumerate(kf.split(X, y)):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    #데이터불균형해소를위하여 SMOTE 적용 (학습 데이터에만)
    smote_tomek = SMOTETomek(random_state=RANDOM_STATE)
    X_resampled_train, y_resampled_train = smote_tomek.fit_resample(X_train, y_train)

    # 리샘플링된 데이터를 리스트에 저장
    resampled_X_train_list.append(X_resampled_train)
    resampled_y_train_list.append(y_resampled_train)
    X_test_list.append(X_test)
    y_test_list.append(y_test)

    # 리샘플링된 데이터를 DataFrame으로 변환
    X_resampled_train_df = pd.DataFrame(X_resampled_train, columns=X.columns)
    y_resampled_train_df = pd.DataFrame(y_resampled_train, columns=['target'])
    X_test_df = pd.DataFrame(X_test, columns=X.columns)
    y_test_df = pd.DataFrame(y_test, columns=['target'])

    # 학습용 데이터 저장
    train_df = pd.concat([X_resampled_train_df, y_resampled_train_df], axis=1)
    train_df.to_csv(f'renew_smote_tomek/train_fold_{fold+1}.csv', index=False)

    # 테스트용 데이터 저장
    test_df = pd.concat([X_test_df, y_test_df], axis=1)
    test_df.to_csv(f'renew_smote_tomek/test_fold_{fold+1}.csv', index=False)
```
